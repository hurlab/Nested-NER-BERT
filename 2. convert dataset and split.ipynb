```python
import pandas as pd
import os

def transform_dataframe_to_single_row(df, df_id):
    # Assuming df has the columns that can be directly mapped or transformed
    # into 'tokens', 'pos_tags', 'chunk_tags', 'ner_tags'.
    # This function will need to be adjusted based on the actual structure of your df.
    
    # Example transformation, real logic depends on your df structure
    single_row = {
        'id': df_id,
        'tokens': df['Token'].tolist(),  
        'ner_tags1': df['NNER_L1'].tolist(),
        'ner_tags2': df['NNER_L2'].tolist(),
        'ner_tags3': df['NNER_L3'].tolist(),
        'ner_tags4': df['NNER_L4'].tolist(),
        'ner_tags5': df['NNER_L5'].tolist(),
        'ner_tags6': df['NNER_L6'].tolist()
    }
    return pd.DataFrame([single_row])
```

```python
# Path to the folder containing your dataframes (assumed to be in CSV format)
train_folder_path = 'Dataset/DATASET_BIONNE_UPDATED/DATASET_BIONNE/en/train_Nested_entity_separated_V4'

# Placeholder for the merged DataFrame
train_merged_df = pd.DataFrame()

# Load and transform each DataFrame
for i, file in enumerate(os.listdir(train_folder_path)):
    if file.endswith(".csv"):  # Check for CSV files
        df_path = os.path.join(train_folder_path, file)
        df = pd.read_csv(df_path)
        df_id, _ = os.path.splitext(file)
        # Transform and merge
        single_row_df = transform_dataframe_to_single_row(df, df_id)
        train_merged_df = pd.concat([train_merged_df, single_row_df], ignore_index=True)

train_merged_df
```

```python
# Path to the folder containing your dataframes (assumed to be in CSV format)
dev_folder_path = 'Dataset/DATASET_BIONNE_UPDATED/DATASET_BIONNE/en/dev_Nested_entity_separated_V4'

# Placeholder for the merged DataFrame
dev_merged_df = pd.DataFrame()

# Load and transform each DataFrame
for i, file in enumerate(os.listdir(dev_folder_path)):
    if file.endswith(".csv"):  # Check for CSV files
        df_path = os.path.join(dev_folder_path, file)
        df = pd.read_csv(df_path)
        df_id, _ = os.path.splitext(file)
        # Transform and merge
        single_row_df = transform_dataframe_to_single_row(df, df_id)
        dev_merged_df = pd.concat([dev_merged_df, single_row_df], ignore_index=True)

dev_merged_df
```

```python
# Merge the two dataframes
train_dev_merged_df = pd.concat([train_merged_df, dev_merged_df], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split

# Assuming dev_merged_df is your dataframe and you want to split it
# Let's split it into 80% train and 20% validation sets as an example

#val_df, test_df = train_test_split(dev_merged_df, test_size=0.5, random_state=42)
train_df, validation_df = train_test_split(train_dev_merged_df, test_size=0.2, random_state=42)

# Now val_df contains 80% of the data and test_df contains 20% of the data
# Resetting the index for both dataframes without adding the old index as a column
train_df = train_df.reset_index(drop=True)
validation_df = validation_df.reset_index(drop=True)
```

```python
from datasets import Dataset
import pandas as pd
import os

# Convert the DataFrames to Hugging Face Dataset format

train_df = Dataset.from_pandas(train_df)
validation_df = Dataset.from_pandas(validation_df)

# If you want to combine them into a single dataset with a split, you can do the following:
from datasets import DatasetDict

BioASQ_NNER_dict = DatasetDict({
    'train': train_df,
    'validation': validation_df,
})
BioASQ_NNER_dict
```
